```python
import numpy as np
```

```python
a = np.array([1,2,3,4,5])
b = np.array([[10,20,30],[40,50,60]])
zeros_arr = np.zeros(3)
ones_arr = np.ones((2,2))
arange_arr = np.arange(0,10)

print("a =", a)
print("b =", b)
print("zeros_arr =", zeros_arr)
print("ones_arr =", ones_arr)
print("arange_arr =", arange_arr)
```

```python
print("Форма массива б:", b.shape)
print("тип данных массива б:", b.dtype)
print("форма массива arange_arr:", arange_arr.shape)
print("тип данных массива arange_arr:", arange_arr.dtype)
```

```python
f_element = a[0]
l_element = a[-1]
b_element = b[0,1]
slice_a = a[1:4]
slice_b = b[:, 1:]

print("f_element =", f_element)
print("l_element =", l_element)
print("b_element =", b_element)
print("slice_a =", slice_a)
print("slice_b =", slice_b)
```
