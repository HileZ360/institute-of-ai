```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.DataFrame({
    "age": [19, 22, 28, 30, 41, 55, 22],
    "events": [3, 1, 4, 6, 10, 12, 2],
    "score": [0.8, 0.7, 0.6, 0.82, 0.9, 0.95, 0.65]
})
df
```

```python
df.isna().mean()
df.duplicated().sum()
df[(df["age"] < 0) | (df["age"] > 100)]
```

```python
plt.figure()
df["age"].hist(bins=5)
plt.xlabel("age")
plt.ylabel("count")
plt.title("Распределение возраста")
plt.show()
```

```python
plt.figure()
plt.boxplot(df["score"])
plt.ylabel("score")
plt.title("Boxplot score")
plt.show()
```

```python
plt.figure()
plt.scatter(df["events"], df["score"])
plt.xlabel("events")
plt.ylabel("score")
plt.title("завис. score от events")
plt.show()
```
